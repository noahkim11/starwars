```python
#filename = 'star wars - new hope.txt'
#location_marker = ['EXT.','INT.']

def parse_script(filename,location_marker):

    f = open(filename,'r')
    change_actor = False
    cur_location = ''
    location_counter = 0
    location = ''
    actor = ''
    lines = ''
    output = []
    
    for row in f:
        #blank line resets location and character change        
        if row == '\n' and change_actor:
            output.append([location, location_counter, actor, lines])
#            print(location, actor, lines)
            change_actor = False
            lines = ''
            continue

        #locations and characters flag
        if row == row.upper() and row != '\n':
            if any([True if x in row else False for x in location_marker]): #if EXT. or INT.        
                location = row.strip('\n')                
            else:
                change_actor = True            
                actor = row.strip('\n')
                lines = ''
            continue
        
        #multiple lines of dialog
        if change_actor:
            lines += row.strip('\n')            

        if not cur_location:
            cur_location = location
            
        if cur_location != location:
#            print(cur_location, location, location_counter)
            
            location_counter += 1
            cur_location = location
            
            
            
#            print(cur_location)
            
        
    f.close()
    return output

```

```python
import pandas as pd

filename = 'star wars - new hope.txt'
location_marker = ['EXT.','INT.']

script = parse_script(filename,location_marker)

df = pd.DataFrame(script, columns=['location', 'location_counter', 'actor', 'lines'])
df = df[df['location']!='']

#for item in script:
#    print(item)

#df.to_csv('output.csv', index=False)
```
